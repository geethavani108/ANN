```python
# First ann
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'C:\\Users\kmmoh\Downloads\Githubupload\churn_modelling.csv')
X=df.iloc[:,3:13]
y = df.iloc[:, 13]
X.head()
```

```python
df.head()
```

```python
geography=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
X=pd.concat([X,geography,gender],axis=1)
X.drop(columns=['Geography','Gender'],axis=1,inplace=True)

X.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size = 0.2, random_state = 0)
#feature scaling
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train,y_train)
X_test=ss.fit_transform(X_test)
```

# Model Building

```python
Initialising the ANN

Defining by adding layers 

Compiling the ANN Train the ANN

Units - number of columns of the dataset

output_dim - number of outputs to be fed to the next layer, if any

activation - activation function which is ReLU in this case
```

```python
#Let us build ann
import keras
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
#Initialising the ANN
ann_classifier=Sequential()
#Defining by adding layers 
ann_classifier.add(Dense(units=6, kernel_initializer='he_uniform',input_dim=11, activation='relu'))
# Adding the second hidden layer
ann_classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))
#Adding final layer
ann_classifier.add(Dense(units=1,kernel_initializer= 'glorot_uniform', activation='sigmoid'))
ann_classifier.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
#fitting the ANN to training set
model_history=ann_classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10, epochs = 150)
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = ann_classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print("Accuracy score       :",score)
```

```python
#Let us build ann
import keras
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU

#Initialising the ANN
classifier=Sequential()
#Defining by adding layers 
classifier.add(Dense(units=10, kernel_initializer='he_normal',input_dim=11, activation='relu'))
classifier.add(Dropout(0.3))
# Adding the second hidden layer
classifier.add(Dense(units = 20, kernel_initializer = 'he_normal',activation='relu'))
classifier.add(Dropout(0.3))
                         
classifier.add(Dense(units = 15, kernel_initializer = 'he_normal',activation='relu'))
classifier.add(Dropout(0.3))
#Adding final layer
classifier.add(Dense(units=1,kernel_initializer= 'glorot_uniform', activation='sigmoid'))
classifier.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
#fitting the ANN to training set
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10, epochs = 100)
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print("Accuracy score       :",score)
```
